```python
from bs4 import BeautifulSoup as soup
import urllib.request as urlReq
```

```python
# Plug in the url

myurl = "http://racing.hkjc.com/racing/info/meeting/Draw/english/Local"

# opening connection
page = urlReq.urlopen(myurl).read()

# creating a soup object
page_soup = soup(page, "html.parser")

```

```python
for div in page_soup.findAll('div'):
    print(div.text)
```
